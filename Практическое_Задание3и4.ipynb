```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

coron = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv', sep=",")
coron = coron.drop(columns=['Province/State', 'Lat', 'Long'])
coron = coron.groupby('Country/Region').sum()

countries = ['Uzbekistan', 'Angola', 'China', 'Japan', 'Poland']

# Создание нового DataFrame, содержащего только выбранные страны
selected_countries = coron.loc[countries]

# Построение графика для каждой страны
for country in countries:
    plt.plot(selected_countries.columns, selected_countries.loc[country], label=country)

# Добавление заголовка и метки осей
plt.title('Статистика COVID-19')
plt.xlabel('Дата')
plt.ylabel('Пострадавших человек')

plt.legend()

plt.show()

```

Группируем данные по странам, суммируя значения для каждой страны.

Выбираем определенные страны к примеру:(‘Uzbekistan’, ‘Angola’, ‘China’, ‘Japan’, ‘Poland’) для анализа и создаем новый DataFrame, содержащий только эти страны.

Строим график для каждой выбранной страны, отображающий количество смертей от COVID-19 в течение времени.

Добавляем заголовок графика и метки осей.И добавляем легенду на график.

Далее добавляем команду для отображения график.

Таким образом, этот код анализирует и визуализирует данные о смертях от COVID-19 для выбранных стран.
